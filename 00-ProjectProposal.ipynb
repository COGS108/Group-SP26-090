## Research Question

Refined Question:

To what extent do college majors predict early-career job insecurity among recent U.S. college graduates (0-2 years post-graduation)? Job insecurity in this project will be analyzed using 3 measurable labor-market outcomes, specifically: (1) job offer attainment and unemployment within two years of graduation, (2) underemployment via role relevance, and (3) starting salary relative to field-specific medians. We will account for key controls known to affect job-market outcomes such as the differences in student interest or competition within the major (proxied by major size/enrollment), academic performance (GPA), prior work experience and prestige (nepotism/university/internship/network). This is primarily a statistical inference related research problem, using group comparisons to estimate the amount of variation in job insecurity explained by major categories. A secondary goal is to determine whether certain fields consistently exhibit/experience higher vulnerability to job insecurity after adjusting for these confounding factors.

## Background and Prior Work

Post graduation, students are thrust into the job market, where they must compete in a space with many competitors to secure a position of employment. While this has been a factor of stress for many students' lives already, as of recently, the job market has been in a worse state, one reason being AI. In Faluyi’s research paper, he cites Murphy on the data of AI,“Artificial intelligence is growing so rapidly that by 2030, AI is expected to contribute to $15.7 trillion to the global economy and permeate most industries (Murphy et al., 2021)” (2). The large contribution that AI makes benefits employers and harms workers, as they are able to work for less costs, which makes them more appealing to hire rather than people. As Faluyi refers to other’s commentary, “Acemoglu and Restrepo (2019) suggest that the jobs most susceptible to automation are often low-skilled and low-paying, which could disproportionately affect workers with limited education and training, potentially exacerbating the wealth gap” (4). Overall, with the tasks that AI is capable of performing, more people are being replaced with lower costs, and that contributes to a lot of stress, including students who are pressured to compete even harder than before.

Of course, even though AI may be a factor in the state of the job market, it may not be the only factor that makes it so competitive. Generally, there may just be a decrease in interest in hirings throughout the years, as the market gets more and more competitive (Ember). As the market gets more competitive, it wouldn’t be inaccurate to say that there is a form of “inflation” where higher level skills shown in interviews or resumes are becoming more appealing and in a sense, required. Students are pressured to achieve these kinds of skills or risk not being as competitive as others in the market, adding a lot of additional stress to their already stressful lives filled with study. Another factor that contributes to differing levels of stress is the differing unemployment rates between majors, which through data, shows certain majors having higher or lower earnings, as well as different unemployment rates (Dickler). So as the market runs as it does currently, less and less people are offered as many opportunities as before.

Our project focuses more on the success of college students divided by major acquiring a position of employment. While AI is not a main focus for our research question, it is still a relevant topic for discussion due to its proximity to the issue we are tackling. This is shown clearly when evaluating the increase in competition generally, regardless of AI's involvement in unemployment rates. Unemployment rates is not a new issue, and that fact is reflected by some of the similar discussions and datasets that have been conducted regarding the issue. In Dickler’s article, it can be seen through graphs and illustrations the differences between majors and their unemployment rates, showing investment into this topic between students of different interests. This is in addition to some of the sources linked below, which include data about the job market and "ghost" jobs. Looking through all of these sources while performing background research, all of this is to say that the job market has reached a point of expectations and levels where those seeking employment are thrust into far too much stress for little value. Thus, our project, which hinges on the unemployment rates of students pursuing certain majors, aims to help alleviate some concerns or at the very least help prepare others for future possibilities.

## Hypothesis


We predict that the college major category will have a strong correlation and be significant at predicting early-career job insecurity, with humanities and social sciences graduates having worse results (higher unemployment, underemployment, and lower starting salaries compared to the median at that occupation) compared to STEM and health related fields. We made this prediction based on the reasoning that STEM and health related fields are more focused on qualifications like experience and skills for hiring, which the degree provides specific skills that are in demand to employers. On the other hand, humanities and social sciences graduates will enter into a broader job market, where their degree has less of a direct connection to the roles they may apply for, leading to an increase in vulnerability to underemployment and earning less than median salary for that role.

## Data

To effectively explore and answer the primary research question and its subquestions, the idealized dataset would consist of longitudinal, individual-level observations across multiple successive annual cohorts (e.g., 2018–2025), following approximately 50,000 U.S. college graduates per year for their first 24 months post-graduation. This idealized structure accounts for macro-economic shocks, such as the rise of AI, structural layoffs, or the residual effect os COVID-19, that create fluctuating demand across sectors. Unlike our current datasets, which force a trade-off between high-level accuracy (NY Fed) and individual-level granularity (Kaggle), this ideal dataset would merge these views into a single, tidy format. By matching the volume of graduates against real-time industry openings ensures an ethical analysis that accounts for uneven subpopulation sizes and market saturation. Additional variables of interest include student GPA, parental socioeconomic status (SES), and debt-to-income ratios. The principal categories of our idealized dataset include: Academic Profile, Socioeconomic Background, Market Dynamics (Supply/Demand), Experience & Access, Job Search, Labor Outcomes, Intergenerational Gap, and Psychological Sentiment.

The idealized principle categories would suffice for a representation of the individual variables and institutional and market metrics necessary to perform a precise statistical inference of the "major effect" while controlling for the most pressing socioeconomic and temporal factors that often confound group averages. A proposed table of variables alongside the rational for our idealized dataset is provided:

|Category|Variables|Reasoning|
|--------|---------|---------|
|Academic Profile|Primary Major/Minor, GPA, University Name/Ranking, Major Cohort Size (Local), Total Institution Size. grads across majors and expected outcomes|Controls for "institutional brand" and internal competition. Large cohort sizes may indicate degree saturation or a crowded local labor market for that specific school's alumni.|
|Socioeconomic Background|Parental Social Economic Status (SES), Student Debt-to-Income Ratio.|Accounts for the "financial cushion" that allows some graduates to wait for better roles, preventing forced underemployment.|
|Market Dynamics (Supply/Demand)|Major Enrollment/Graduation Volume, Sector-specific Job Openings, Supply-to-Demand Ratio, Major Unemployment and Underemployment.|Normalizes comparisons between over-saturated fields and high-demand niches to avoid bias from uneven population sizes.|
|Experience & Access|Number/Prestige of Internships, Prior Work Experience, Referral Status (Nepotism).|Isolates whether a job was secured through specific skills provided by the degree or through existing social networks.|
|Job Search|Method of hire, Number of interviews, Time-to-offer (weeks).|Measures the ease of obtaining a job, a key indicator of market demand for specific majors.|
|Labor Outcomes|Initial Salary, Job Title, SOC Code (Role Relevance), Benefits Package.|Allows for precise calculation of underemployment by comparing job duties (SOC) directly to the major’s curriculum. Relevant for the perceived well-being of new individuals.|
|Intergenerational Gap|Parental Income (at same age).|Measures whether a major choice facilitates upward mobility or results in economic stagnation relative to the previous generation.|
|Psychological Sentiment|Perceived Job Security (1-10), Search Fatigue, Job Satisfaction.|Captures the subjective experience of job insecurity and vulnerability that financial metrics alone might miss.|

Early/Possible Data:

1. Job Board Effectiveness for College Students (https://www.kaggle.com/datasets/likithagedipudi/job-board-effectiveness-for-college-students)

Description: A large-scale study of how entry-level candidates fare on major hiring platforms like LinkedIn, Handshake, and Indeed. This dataset features 100,000 records of individual student data, including demographics (major, region), academic stats (GPA, internships), and a full "application funnel" that tracks progress from the first application to the final job offer.


2. The labor market for recent college grads, New York Fed - Outcomes by Major 2024 (https://www.newyorkfed.org/research/college-labor-market#--:explore:outcomes-by-major)

Description: An annual and quarterly report based on U.S. Census Bureau data that analyzes the employment experiences of recent college graduates (ages 22–27). This dataset contains aggregated statistics on unemployment, underemployment, and median wages specifically broken down by college major. It also includes "mid-career" benchmarks and graduate degree shares for broader context.


3. Bureau of labor statistics, JOLTS (https://www.bls.gov/jlt/)

Description: A Bureau of Labor Statistics (BLS) program that tracks the overall "health" and "churn" of the U.S. job market. This datasource provides monthly and annual estimates of job openings, hires, and separations (quits and layoffs) at both the national and state level across different industry sectors.

The primary disparity between our idealized dataset and the real-world lies in the unit of observation and data integration. Our real-world data from the NY Fed provides accurate aggregated outcomes but lacks individual outcomes data. Conversely, the Kaggle dataset offers individual-level metrics but doesn't capture the broader macroeconomic supply/demand ratios provided by the Bureau of Labor Statistics/JOLTS. In an ideal world, we would see exactly how a single Philosophy major with a 3.8 GPA navigates a market where the graduate-to-opening ratio is, for instance, 5:1. Since our actual data is siloed, we will bridge this gap by using the JOLTS industry-level demand to weight the outcomes we see in the NY Fed and Kaggle datasets, ensuring our conclusions about job insecurity are adjusted for the actual number of graduates entering each field.
